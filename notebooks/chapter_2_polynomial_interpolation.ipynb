## Example 2.1

Try to interpolate the data from 


$x_i$| $0$ | $1$ | $2/3$
--- | --- | --- | ---
$y_i$| $1$ | $0$ | $0.5$

by means of the polynomial function: $P_2(x) = a_2 x^2 + a_1 x = a_0 $

```python
# Define Vandermonde matrix and right hand side vector
x = np.array([[1,0,0],[1,1,1],[1,2/3,4/9]])
y = np.array([1,0,1/2])

# Solve the linear system of equations
a = np.linalg.solve(x,y)

# Calculate the polynomial
t = np.linspace(0,1)
p2 = a[0] + a[1] *t + a[2]*t**2

# Plot datapoints
plt.plot([0,1,2/3],[1,0,1/2], "o")

# Plot solution
print(f"The solution vector is: {a}")
plt.plot(p2, t)

# Plot analytical solution
fx = np.cos(np.pi/2 *t)
plt.plot(t, fx, "--k")

plt.grid()
```
